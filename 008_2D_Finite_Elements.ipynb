```python
import numpy as np
import matplotlib.tri as mptri
import matplotlib.pyplot as plt
from netgen.geom2d import *
from ngsolve import *
import os, meshio
import quadpy
from sympy import simplify, lambdify, symbols, Symbol, init_printing, integrate, diff, Array, Matrix
plt.rc("text", usetex=True)
meshDir = os.path.join(os.getcwd(), "meshes2D")
resultsDir = os.path.join(os.getcwd(), "results")
os.makedirs(meshDir, exist_ok = True)
os.makedirs(resultsDir, exist_ok = True)
```

```python
class basis():
    def __init__(self, deg, basis_type):
        deg = int(deg)
        if basis_type == 'L':          # 1D Lagrange basis of degree `deg`
            z=Symbol('z')
            Xi=np.linspace(-1,1,deg+1)
            def lag_basis(k):
                n = 1.
                for i in range(len(Xi)):
                    if k != i:
                        n *= (z-Xi[i])/(Xi[k]-Xi[i])
                return n
            N = Array([simplify(lag_basis(m)) for m in range(deg+1)])            
            dfN = diff(N,z)+1.e-25*N
            self.Ns=lambdify(z, N, "numpy")
            self.dN=lambdify(z, dfN, "numpy")
        
        elif basis_type == "T": # 2D lagrange basis of degree `deg`. Only supported `deg` = 1, 2
            xi, eta = symbols("xi, eta", real=True)
            
            N = Array([1+1.e-19*xi, eta, 1-xi-eta])
            
```

```python
def generateNetgenMesh(hval):
    geo = SplineGeometry()

    rect = geo.AddRectangle((-0.1, -0.1), (0.1, 0.1), leftdomain=1, rightdomain=0, bc="outer")
    circ = geo.AddCircle((0, 0), 0.025, leftdomain=0, rightdomain=1, bc = "inner")

    msh = geo.GenerateMesh(maxh=hval)

    points = []
    for pt in msh.Points():
        points.append([*pt])

    points = np.array(points)

    points = points[:, :2]

    el = []
    for elem in msh.Elements2D():
        el.append([int(str(i)) for i in elem.points])

    el = np.array(el, int)
    return msh, points, el-1
```
